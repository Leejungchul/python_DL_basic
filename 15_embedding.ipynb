```python
sentense = [
    'I love my dog',
    'I love my cat',
    'You love my dog',
    'Do you think my dog is amazing?'
]
```

### 토큰화

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical   # 원핫인코딩 방식
```

```python
tokenize = Tokenizer(num_words=100,  # 단어 100개로 토큰화
oov_token = '<oov>')  # 단어가 단어 100개 중에 없을 때 표시

tokenize.fit_on_texts(sentense)
```

```python
tokenize.word_index   # 숫자가 적을 수록 자주 나온 단어
```

```python
# 시퀀스로 변환

sentense = tokenize.texts_to_sequences(sentense)
sentense
```

```python
test_sen = [
    'I really love my dog',
    'my dog loves my friend',
    'do you think my dog is amazing'
]
```

```python
test_sen = tokenize.texts_to_sequences(test_sen)
```

#### 문장마다 길이가 달라서 길이를 맞춰줘야함
#### 패딩 설정 : rnn에서 타임스탭을 맞추는 과정

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

pad_sequence = pad_sequences(test_sen, maxlen = 7, padding='pre')   # (문장, maxlen: 맞춰줄 길이, padding : - pre : 앞쪽에 0 채우기/버리기, -post : 뒤에 0 채우기/버리기)
```

### 임베딩(원핫인코딩)

```python
from tensorflow.keras.utils import to_categorical

sentence = to_categorical(pad_sequence)
```

```python
sentence   # 단어의 개수만큼 원핫이 생김 -> 11개
```

### 임베딩
- 원핫 하면 데이터가 너무 길어지고 0이 많아짐




```python
paper = ['많은 것을 바꾸고 싶다면 많은 것을 받아들여라']

tokenizer = Tokenizer()
tokenizer.fit_on_texts(paper)
```

```python
tokenizer.word_index
```

```python
idx_paper = tokenizer.texts_to_sequences(paper)
idx_paper
```

```python
to_categorical(idx_paper)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

model = Sequential()
model.add(Embedding(input_dim=6, output_dim=3))   # 6개 데이터 인풋, 출력 3개

model.compile(optimizer='adam', loss='mse')

    
embedding = model.predict(idx_paper)


```
