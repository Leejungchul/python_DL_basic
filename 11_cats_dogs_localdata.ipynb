```python
cat_X = []
cat_y = []
```

```python
import os
cat_list = os.listdir('/content/cats_and_dogs_filtered/train/cats')
```

```python
len(cat_list)
```

```python
for cat in cat_list:
  im = Image.open('/content/cats_and_dogs_filtered/train/cats/'+cat)
  new_image = im.resize((32,32))
  img = np.array(new_image)
  cat_X.append(img)
  cat_y.append(0)



```

```python
import os
dog_list = os.listdir('/content/cats_and_dogs_filtered/train/dogs')
```

```python
dog_X = []
dog_y = []
```

```python
dog_list
```

```python
for dog in dog_list:
  im = Image.open('/content/cats_and_dogs_filtered/train/dogs/'+dog)
  new_image = im.resize((32,32))
  img = np.array(new_image)
  dog_X.append(img)
  dog_y.append(1)
```

```python
X = cat_X + dog_X
```

```python
y = cat_y + dog_y
```

```python
X = np.array(X)
```

```python
y=np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1313, shuffle=True, test_size=0.2)
```

```python
X_train,X_val, y_train, y_val = train_test_split(X_train, y_train, random_state=1313, shuffle=True,test_size=0.2)
```

```python
X_val = X_val / 255.
```

```python
X_train = X_train / 255.
X_test = X_test / 255.
```

```python
len(y)
```

```python
y_test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential()
model.add(layers.Conv2D(32,(3,3), padding='same', input_shape=(32,32,3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Conv2D(64,(3,3), padding='same', activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2,2)))

model.add(layers.Conv2D(64,(3,3), padding='same', activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(128,(3,3), padding='same', activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2,2)))
model.add(layers.Dropout(0.25))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics = ['acc'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=30, batch_size = 128, validation_data = (X_val, y_val))
```

```python
model.evaluate(X_test,y_test)
```
